```python
import pandas as pd
import csv

def read_csv_with_error_handling(filepath):
    data = []
    with open(filepath, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)  # Read the header
        for row in reader:
            try:
                # Try to process the row as needed (convert to appropriate types, etc.)
                data.append(row)
            except Exception as e:
                print(f"Skipping row due to error: {e}")
    return pd.DataFrame(data, columns=headers)

df = read_csv_with_error_handling('/content/house_prices.csv')

df.head()
```

```python
df.info()
```

```python
#EDA
# Remove rows with missing values
df_cleaned = df.dropna(axis=0)

```

```python
#Reomoving any duplicated rows
df_final=df_cleaned.dropna()
```

```python
#Renaming the features to a specified format
df_final.rename(columns={'Amount(in rupees)': 'Amount', 'Price (in rupees)': 'Price'}, inplace=True)
```

covert the Amount to float type after computing lac,cr to whote number format

```python
import re
import numpy as np

def convert_amount(amount):
    if pd.isnull(amount):
        return np.nan

    amount = amount.replace(',', '')

    if 'Lac' in amount:
        value = float(amount.replace('Lac', '').strip())
        return value * 100000
    elif 'Cr' in amount:
        value = float(amount.replace('Cr', '').strip())
        return value * 10000000
    # Handle cases like 'Call for Price'
    elif not amount.replace('.', '', 1).isdigit():
        return np.nan
    else:
        return float(amount)

df_final['Amount'] = df_final['Amount'].apply(convert_amount)
```

```python
#convert the Price to float type
# Handle empty strings in 'Price' column before converting to float
df_final['Price'] = df_final['Price'].replace('', np.nan)
df_final['Price'] = df_final['Price'].astype(float)
```

```python
df_final.head()
```

```python
df_final['Price'].fillna(df_final['Price'].mean(), inplace=True)
```

Here “sweetviz.analyze()” is just a function that analyzes the data frame and we have to pass the target feature as a parameter. Always remember the parameters should be in form of a list. You can check out the documentation and play with different parameters.

```python
df_final['Price'] = pd.to_numeric(df_final['Price'], errors='coerce')

my_report = sv.analyze([df_final,"Train"],target_feat='Price')
my_report.show_html('report.html')
```

Whenever you will hover on a particular feature, it will show full detail about that feature.
